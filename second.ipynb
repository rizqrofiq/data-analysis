# Feature Scaling with Python

Feature scaling is a method used to normalize the range of independent variables or features of data. In data processing, it is also known as data normalization and is generally performed during the data preprocessing step.

```python
import pandas as pd
import numpy as np

from pandas import DataFrame
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
```

```python
df = pd.read_csv("./dataset/result.csv")
df
```

```python
df.describe()
```

```python
df.loc[2, "Jum_kamar"] = 100
df
```

```python
feature_minmax = df.copy()
col_minmax = ["Jum_kamar"]

features = feature_minmax[col_minmax]
min_max = MinMaxScaler()

kamar_minmax = min_max.fit_transform(features.values)

kamar_minmax
```

```python
feature_minmax[col_minmax] = kamar_minmax
feature_minmax
```

```python
feature_minmax.to_csv("./dataset/apartment_minmax.csv", index=False)
```
