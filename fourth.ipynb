```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./dataset/building_permits.csv")
df
```

```python
df_missing = df.isna().sum()
df_missing[df_missing > 0]
```

## Task #1

```python

df["Zipcode"].fillna(1111, inplace=True)
df["Zipcode"].isna().sum()
```

```python
df[df["Zipcode"] == 1111]
```

## Task #2

```python
import time, datetime

def convert_to_timestamp(x):
    return datetime.datetime.strptime(x, "%m/%d/%Y").timestamp()

def convert_to_datestring(x):
    return datetime.datetime.fromtimestamp(x).strftime("%m/%d/%Y")
```

```python
modified_complete_date = df["Completed Date"].dropna().map(convert_to_timestamp).sort_values(ascending=True)
```

```python
oldest_date = convert_to_datestring(list(modified_complete_date)[0])
```

```python
df["Completed Date"].fillna(oldest_date, inplace=True)
df["Completed Date"]
```

```python
df[df["Completed Date"] == oldest_date]
```

## Task #3

```python
df.dropna(subset="Proposed Use", inplace=True)
df["Proposed Use"].isna().sum()
```
