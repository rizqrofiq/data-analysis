```python
import random
from scipy.stats import ttest_ind, ttest_rel
from scipy import stats

df = df[(np.abs(stats.zscore(df[" loan_status"])) < 3)]

term_0 = df[df[' loan_status'] == "Approved"][' loan_term']
term_1 = df[df[' loan_status'] == "Rejected"][' loan_term']


term_0 = term_0.values.tolist()
term_0 = random.sample(term_0, 10)
term_1 = term_1.values.tolist()
term_1 = random.sample(term_1, 10)

ttest, pval = ttest_ind(term_1, term_0, equal_var=False)
print("ttest", ttest)
print('p value', pval)

if pval < 0.05:
    print("Ho Ditolak")
else:
    print("Ho Diterima")
```

```python
import random
from scipy.stats import ttest_ind
from scipy import stats

df = df[(np.abs(stats.zscore(df["Active"])) < 3)]

income_1 = df[df['Deaths'] == 1]['Active']
income_0 = df[df['Deaths'] == 2]['Active']

income_0 = income_0.values.tolist()

if len(income_0) >= 10:
    income_0 = random.sample(income_0, 10)
else:
    income_0 = income_0

income_1 = income_1.values.tolist()

if len(income_1) >= 10:
    income_1 = random.sample(income_1, 10)
else:
    income_1 = income_1

ttest, pval = ttest_ind(income_1, income_0, equal_var=False)
print("ttest", ttest)
print('p value', pval)

if pval < 0.05 or np.isnan(pval):
    print("H0 ditolak")
else:
    print("H0 diterima")

```

```python
# contigency table
c_t = pd.crosstab(df[' education'], df[' loan_status'], margins = False)
c_t
```

```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2

stat, p, dof, expected = chi2_contingency(c_t)
print('dof=%d' % dof)
print('p_value', p)
print(expected)

# interpret test-statistic
prob = 0.99
critical = chi2.ppf(prob, dof)
print('probabilitas=%.3f, critical=%.3f, stat=%.3f' % (prob, critical, stat))
if abs(stat) >= critical:
  print('Dependent (tolak H0)')
else:
  print('Independent (terima H0)')
```
