```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# This would hide FutureWarning from console
warnings.simplefilter(action="ignore", category=FutureWarning)
```

## Dataset: House Ownership

A Real Estate Ownership Classification Task

```python
df = pd.read_csv("./dataset/Ownership.csv")
df.head()
```

## 1. Hitung data unik

```python
selected_columns = ["Education Level", "Gender", "Marital Status", "Home Ownership"]
for column in df:
    if column in selected_columns:
        value_count = df[column].value_counts()
        print(value_count, "\n")
```

## 3. Tampilkan semua tipe data pada semua kolom

```python
df.dtypes
```

## 4. Tampilkan deskripsi untuk kolom  yang tipe datanya numerical

```python
numerical_columns = ["Num Bathrooms", "Num Bedrooms", "Num Cars", "Num Children", "Num TVs"]

df[numerical_columns].describe()
```

## 5. Coutplot untuk masing-masing kolom maks 5 (berikan style)

```python
df_selected = df[selected_columns]
sns.set_style('whitegrid')

fig, ax = plt.subplots(4,1, figsize=(16, 20))
for i, subplots in zip(df_selected, ax.flatten()):
    sns.countplot(df_selected, x=i, ax=subplots, palette="Set2")

plt.show()
```
