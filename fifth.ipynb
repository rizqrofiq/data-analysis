```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Dataset: Gender Classification

```python
df = pd.read_csv("./dataset/gender_classification.csv")
df.head()
```

## Todo

1. Menghitung data unik pada setiap kolom
2. Countplot untuk masing-masing koloom
3. Countplot untuk masing-masing kolom fitur berdasarkan gender

### 1. Menghitung data unik pada setiap kolom

```python
for column in df:
    # print(df[column].unique())
    # print(f"Variant {column}: {df[column].unique()}")
    value_count = df[column].value_counts()
    print(value_count, "\n")
```

### 2. Countplot untuk masing-masing koloom

```python
fig, ax = plt.subplots(5,1, figsize=(10,20))
for i, subplots in zip(df, ax.flatten()):
    sns.countplot(df, x=i, ax=subplots)

plt.show()
```

### 3. Countplot untuk masing-masing kolom fitur berdasarkan gender

```python
fig, ax = plt.subplots(5,1, figsize=(10,20))
for i, subplots in zip(df.drop(["Gender"], axis=1), ax.flatten()):
    sns.countplot(df, x=i, ax=subplots, hue=df["Gender"])

plt.show()
```
